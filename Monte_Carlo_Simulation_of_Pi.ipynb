Monte Carlo Simulation with Pie

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10000
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

# Check if points are inside the circle
inside = x**2 + y**2 <= 1

# Estimate Pi
pi_estimate = 4 * np.sum(inside) / n

# Plot the results
plt.figure(figsize=(6,6))
plt.scatter(x[inside], y[inside], color='g', s=2, label='Inside Circle')
plt.scatter(x[~inside], y[~inside], color='red', s=2, label='Outside Circle')

# Draw the circle
circle = plt.Circle((0, 0), 1, color='blue', fill=False)
plt.gca().add_artist(circle)

plt.title(f"Monte Carlo PI Simulation (Estimated PI: {pi_estimate:.6f})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis('equal')
plt.legend()
plt.grid(False)
plt.show()
```
